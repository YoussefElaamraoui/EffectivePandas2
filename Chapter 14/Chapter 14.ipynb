# 1 - Convert a column with date information to a date

```python
# Does not enforce utc
date_time = pd.to_datetime(dates_series)
```

```python
date_time
```

# 2 - Convert a date column into UTC dates  

```python
utc  = pd.to_datetime(dates_series,utc=True)
```

```python
utc
```

# 3 - Convert a date column into local dates with a time zone

```python
time_zone =  utc.dt.tz_convert("Europe/Rome")
```

```python
time_zone
```

# 4 - Convert a date column into epoch values 

```python
# Epoch time (also called Unix timestamp) is the number of seconds since January 1, 1970 (UTC).
seconds = time_zone.astype(int) // 10**9 
```

```python
seconds
```

# 5 - Convert an epoch number into UTC. 

```python
reverse_utc = pd.to_datetime(seconds,unit='s').dt.tz_localize('UTC')
```
