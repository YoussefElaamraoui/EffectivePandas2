```python
import numpy as np
import pandas as pd 
```

```python
url = 'https://github.com/mattharrison/datasets/raw/master/data/alta-noaa-1980-2019.csv'
alta_df = pd.read_csv(url,engine='pyarrow',dtype_backend='pyarrow')
dates = pd.to_datetime(alta_df.DATE)
```

```python
snow = alta_df.SNOW.rename(dates)
```

# 1 - Create a histogram from a numeric column. Change the bin size 

```python
snow.plot.hist(title='Snowfall',bins=20)
```

```python
# excluding the day with zero falls
snow[snow>0].plot.hist(title='Snowfall',bins=20)
```

```python
snow[snow>10].plot.hist(title='Snowfall',bins=20)
```

```python
snow.plot.hist(title='Snowfall',bins=30)
```

# 2 - Create a boxplot from a numeric column 

```python
# most of the time it does not snow so when i does it shows as outlier
#this is why there are so many dot
snow.plot.box() 
```

```python
snow[lambda snow: snow>0].plot.box()
```

# 3 - Create a Kernel Density Estimate plot from a numeric column  

```python
snow[lambda snow: snow>0].plot.kde()
```

# 4 - Create a line from a numeri column 

```python
snow.plot.line()
```

# 5 - Create a bar plot from the frequency count of a categorical column

```python
snow.resample('ME').sum().plot.barh()
```
